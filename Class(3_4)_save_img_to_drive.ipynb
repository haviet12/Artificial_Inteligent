```python
from keras.datasets import cifar100
import matplotlib.pyplot as plt
(x_train, y_train),(x_test,y_test)=cifar100.load_data()

for i in range (30):
  plt.tight_layout
  plt.subplot(5,6,i+1)
  plt.imshow(x_train[i], cmap='gray', interpolation='none')
```

```python
x_train =x_train.reshape(50000, 32*32*3)
x_train.astype('float32')/255.0
x_test =x_test.reshape(10000, 32*32*3)
x_test.astype('float32')/255.0

from keras.utils import to_categorical
y_train= to_categorical(y_train, 100)
y_test= to_categorical(y_test, 100)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model= Sequential()
model.add(Dense(3000, activation='relu', input_shape=(32*32*3,), name='input'))
model.add(Dense(1000, activation='relu', name='hiden_1'))
# model.add(Dense(2000, activation='relu',  name='hiden_2'))
# model.add(Dense(1000, activation='relu',  name='hiden_3'))
model.add(Dense(100, activation='softmax',  name='output'))
model.summary()

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1, batch_size=128)
t_loss, t_accur=model.evaluate(x_test, y_test)
```
