**mtcar**

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
url = 'mtcars.csv'
data = pd.read_csv(url)

data.hist() # tinh histogram 
```

```python
corrlaration= data.corr() # xet su tuong quan 
print(corrlaration)
```

```python
import seaborn as sns
sns.heatmap(corrlaration, cmap='magma', linecolor='white', linewidths=0.5, annot=True)
```

```python
data.plot(kind='scatter',x='wt', y='mpg', figsize=(9,9), color ='green')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression() # tao 1 mo hinh 
# ham tuyen tinh ax+b
model.fit(X= pd.DataFrame(data['wt']), y= data['mpg'])  # quy trinh hoc va tu dieu chinh cho may 
print(model.intercept_) # hang so b
print(model.coef_) # hang so a
model.score(X= pd.DataFrame(data['wt']), y= data['mpg']) # do chinh xac ( tinh theo %)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color='black', xlim = (0,7))
y_prid = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_prid, color = 'blue')
```

```python
## ham bac 2
model_2= linear_model.LinearRegression() ## tao 1 mo hinh moi
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model_2.fit(X,y=data['mpg']) # cho hoc may va tu dieu chinh gia tri
print(' gia tri so a: ', model_2.coef_)
print(' gia tri so b:', model_2.intercept_)
print(X)
print('do chinh xac la:', model_2.score(X,y=data['mpg']))


data.plot(kind='scatter', x='wt', y='mpg',figsize = (9,9), color='black', xlim = (data['wt'].min()-1,data['wt'].max()+1))
range= np.arange(data['wt'].min(), data['wt'].max(), 0.1)# xap xep lai cac gia tri cua data['wt] tu be den lon de ve do thi

X_new= pd.DataFrame([range, range**2]).T



y_pre=model_2.predict(X=X_new) ## tinh hoi quy cho y dua vao X 
plt.plot(range,y_pre, 'green')
```

```python
####### ham bac 3
model_3 = linear_model.LinearRegression()
X=pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
model_3.fit(X, y=data['mpg'])
print(' he so a:', model_3.coef_)
print('he so b:', model_3.intercept_)
print(' do chinh xac:', model_3.score(X,y=data['mpg']))

data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='red')
range = np.arange(data['wt'].min(), data['wt'].max(),0.1)
X_new= pd.DataFrame([range, range**2, range**3]).T
y_pred=model_3.predict(X_new)
plt.plot(range, y_pred, 'blue')
```

```python
### ham bac 15
model_15= linear_model.LinearRegression()
x=pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4,data['wt']**5,
                data['wt']**6,data['wt']**7,data['wt']**8,data['wt']**9,data['wt']**10,
                data['wt']**11,data['wt']**12,data['wt']**13,data['wt']**14,data['wt']**15]).T
model_15.fit(x, y=data['mpg'])
print('he so a:', model_15.coef_)
print('he so b:', model_15.intercept_)
print('do chinh xac:', model_15.score(x, y=data['mpg']))

data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black')
range = np.arange(data['wt'].min(), data['wt'].max(), 0.1)
X_new= pd.DataFrame([range, range**2, range**3, range**4, range**5
                     , range**6, range**7, range**8, range**9, range**10
                     , range**11, range**12, range**13, range**14, range**15]).T
y_pred= model_15.predict(X_new)
plt.plot(range, y_pred, 'green')
```

```python
########  HỒI QUY VỚI 2 BIẾN ĐẦU VÀO   ###############
model= linear_model.LinearRegression()

model.fit(X = data.loc[:, ['wt', 'hp']], y = data['mpg'])
print('he so a: ', model.coef_)
print('he so b: ', model.intercept_)
print('do chinh xac:', model.score(X = data.loc[:, ['wt', 'hp']],y= data['mpg']))
# mpg = -3,877*wt - 0,031*hp + 37,22
```

Advertising


```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
```

```python
url = 'Advertising.csv'
data= pd.read_csv(url)
data.hist()
plt.show()
sns.pairplot(data)
```

```python
## xac dinh su tuong quan 
corralation= data.corr()
print(corralation)
```

```python
sns.heatmap(corralation, annot=True)
```

```python
data.plot(kind='scatter', x='TV', y='sales', xlim=(data['TV'].min(),data['TV'].max()+1), figsize=(9,9), color='green')
```

```python
model= linear_model.LinearRegression()
X = pd.DataFrame([data['TV'], data['TV']**2, data['TV']**3, data['TV']**4, data['TV']**5]).T
model.fit(X, y=data['sales'])
print('he so a:', model.coef_)
print('he so b:', model.intercept_)
print('do chinh xac:', model.score(X, y=data['sales']))
```

```python
range= np.arange(data['TV'].min(), data['TV'].max(),0.1)
X_new= pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred= model.predict(X_new)
data.plot(kind='scatter', x='TV', y='sales', xlim=(data['TV'].min(),data['TV'].max()+1), figsize=(9,9), color='green')
plt.plot(range, y_pred, 'red')
```

50_Startup


```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
url='50_Startups.csv'
data= pd.read_csv(url)
data.hist()
sns.pairplot(data)
plt.show()
```

```python
corralation= data.corr()
sns.heatmap(corralation, annot=True)
data.plot(kind='scatter', x='Administration', y='Profit', figsize=(9,9), xlim=(data['Administration'].min(), data['Administration'].max()),color="red")
```

```python
x = pd.DataFrame([data['Administration'],data['Administration']**2,data['Administration']**3,data['Administration']**4,data['Administration']**5]).T
model= linear_model.LinearRegression()
model.fit(x, y=data['Profit'])
print('he so a:', model.coef_)
print('he so b:', model.intercept_)
print('do chinh xac:', model.score(x, y=data['Profit']))
```

```python
data.plot(kind='scatter', x='Administration', y='Profit', figsize=(9,9), xlim=(data['Administration'].min(), data['Administration'].max()),color="red")
range=np.arange(data['Administration'].min(),data['Administration'].max(),0.1)
X_new= pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred= model.predict(X_new)
plt.plot(range, y_pred, color='green')
```
