```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
filename='/content/drive/MyDrive/MyPicture/pic.jpg'
img3 =plt.imread(filename)
detector =MTCNN()
faces= detector.detect_faces(img3)
plt.imshow(img3)

ax =plt.gca()
for face in faces:
  x,y,width,heigh= face['box']
  rect=plt.Rectangle((x,y),width,heigh,fill = False, color ='red')
  ax.add_patch(rect)
  
  # hien cac diem keypoint tren khuon mat
  for key,value in face['keypoints'].items():
    dot =plt.Circle(value,radius=2, color ='red')
    ax.add_patch(dot)
  plt.savefig('face.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN

def draw_faces(filename, result_list):
  data = plt.imread(filename)
  for i in range(len(result_list)):
    x1,y1, width, heigh = result_list[i]['box']
    x2 = x1+ width
    y2 = y1+ heigh
    plt.subplot(1, len(result_list),i+1)
    plt.axis('off')
    plt.imshow(data[y1:y2,x1:x2])
  plt.show()
```

```python
filename ='/content/drive/MyDrive/MyPicture/pic.jpg'
img = plt.imread(filename)
detector =MTCNN()
faces=detector.detect_faces(img)
draw_faces(filename, faces)
for i in range(len(faces)):
  path = 'single_face'+ str(i)+'.jpg'
  print(i)
  plt.savefig('path')
```

```python
## using RetNet
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

model = ResNet50(weights='imagenet')

img_path = '/content/drive/MyDrive/MyPicture/bigpull.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

```python
## using vgg 16
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np

model = VGG16(weights='imagenet', include_top=True)

img_path = '/content/drive/MyDrive/MyPicture/bigpull.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

features = model.predict(x)
print('Predicted:', decode_predictions(features, top=3)[0])
```
